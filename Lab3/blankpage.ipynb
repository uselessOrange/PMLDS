```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('exercise_dataset.csv')

print(data.describe)
print(data.head)

print(type(data))
```

```python

first_5_data = data.head(5)

print(first_5_data)
```

```python


# Select the first 5 entries and relevant features
first_5_entries = data.head(5)
features_to_plot = ['Calories Burn', 'Actual Weight']

# Plot a bar plot for the selected features in the first 5 entries
first_5_entries[features_to_plot].plot(kind='bar', figsize=(10, 6))

# Set plot labels and title
plt.xlabel('Entry')
plt.ylabel('Values')
plt.title('Bar Plot for Calories Burn and Actual Weight (First 5 entries)')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability

plt.show()
```

```python


# Group by 'Exercise' and calculate the mean 'Calories Burn'
average_calories_by_exercise = data.groupby('Exercise')['Calories Burn'].mean()

# Plot a bar plot for average calories burned vs Exercise
plt.figure(figsize=(12, 6))
average_calories_by_exercise.plot(kind='bar', color='skyblue')
plt.xlabel('Exercise')
plt.ylabel('Average Calories Burned')
plt.title('Average Calories Burned vs Exercise')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
