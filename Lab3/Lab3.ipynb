Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
```

Loading Data

```python
data = pd.read_csv('exercise_dataset.csv')
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Display summary statistics of the dataset
print("\nSummary statistics of the dataset:")
print(data.describe())
```

Histogram showing the number of people of a specific age

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Age'], bins=15, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.title('Histogram of Ages')
plt.show()
```

Bar graph

```python
# Select the first 5 entries and relevant features
first_5_entries = data.head(5)
features_to_plot = ['Calories Burn', 'Actual Weight']

# Plot a bar plot for the selected features in the first 5 entries
first_5_entries[features_to_plot].plot(kind='bar', figsize=(10, 6))

# Set plot labels and title
plt.xlabel('Entry')
plt.ylabel('Values')
plt.title('Bar Plot for Calories Burn and Actual Weight (First 5 entries)')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability

plt.show()
```

Bar graph #2

```python
# Group by 'Exercise' and calculate the mean 'Calories Burn'
average_calories_by_exercise = data.groupby('Exercise')['Calories Burn'].mean()

# Plot a bar plot for average calories burned vs Exercise
plt.figure(figsize=(12, 6))
average_calories_by_exercise.plot(kind='bar', color='green')
plt.xlabel('Exercise')
plt.ylabel('Average Calories Burned')
plt.title('Average Calories Burned vs Exercise')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

Scatterplot showing Dream Weight vs Actual Weight

```python

plt.figure(figsize=(10, 6))
plt.scatter(data['Actual Weight'], data['Dream Weight'], color='purple', alpha=0.7)
plt.xlabel('Dream Weight')
plt.ylabel('Actual Weight')
plt.title('Scatterplot of Dream Weight vs Actual Weight')
plt.show()
```

Boxplot (example using some random data)

```python

plt.figure(figsize=(10, 6))
plt.boxplot([data['Age'],data['Actual Weight']])
plt.xlabel('Age and Actual Weight')
plt.ylabel('Values')
plt.title('Boxplot')
plt.show()
```

```python
sns.catplot(data=data, x="Exercise", y="Calories Burn", hue="Gender", kind="swarm")
```
